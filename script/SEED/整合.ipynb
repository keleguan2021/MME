```python
# Shuffle
for i in range(15):
    print("Subject:" + str(i))
    for j in range(3):
        print("  Session:" + str(j))
        for k in range(15):
            temp_feature_map = np.load("trials/feature/subject_" + str(i) + "/section_" + str(j) + "/trial_" + str(k) + ".npy")
            temp_temporal_map = np.load("trials/temporal/subject_" + str(i) + "/section_" + str(j) + "/trial_" + str(k) + ".npy")
            temp_label = np.array([origin_label[k] for _ in range(temp_feature_map.shape[0])])
            if(k == 0):
                train_feature_map = temp_feature_map
                train_temporal_map = temp_temporal_map
                train_label = temp_label
            else:
                train_feature_map = np.concatenate((train_feature_map, temp_feature_map), axis=0)
                train_temporal_map = np.concatenate((train_temporal_map, temp_temporal_map), axis=0)
                train_label = np.concatenate((train_label, temp_label), axis=0)
   
        index = np.arange(train_feature_map.shape[0])
        np.random.shuffle(index)

        train_feature_map = train_feature_map[index]
        train_temporal_map = train_temporal_map[index]
        train_label = train_label[index]

        test_feature_map = train_feature_map[2035:]
        test_temporal_map = train_temporal_map[2035:]
        test_label = train_label[2035:]
        
        train_feature_map = train_feature_map[:2035]
        train_temporal_map = train_temporal_map[:2035]
        train_label = train_label[:2035]

        np.save("trials/train/feature/subject_" + str(i) + "/section_" + str(j) + "_label.npy", train_label)
        np.save("trials/train/temporal/subject_" + str(i) + "/section_" + str(j) + "_data.npy", train_temporal_map)
        np.save("trials/train/feature/subject_" + str(i) + "/section_" + str(j) + "_data.npy", train_feature_map)

        np.save("trials/test/feature/subject_" + str(i) + "/section_" + str(j) + "_label.npy", test_label)
        np.save("trials/test/temporal/subject_" + str(i) + "/section_" + str(j) + "_data.npy", test_temporal_map)
        np.save("trials/test/feature/subject_" + str(i) + "/section_" + str(j) + "_data.npy", test_feature_map)
```
